```python
import numpy as np
from scipy.stats import norm
```

```python
mu = 0.4
sigma = 0.5
dt = 1 / 252
S0 = 1
n = 200
```

```python
T = np.linspace(0, (n-1)* dt, n)
```

```python
W = np.zeros(n)
```

```python
W[1:] = np.cumsum(norm.rvs(size=n-1) * np.sqrt(dt))
```

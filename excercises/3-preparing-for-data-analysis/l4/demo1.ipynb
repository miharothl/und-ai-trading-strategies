# Demo

Using matplotlib to plot stock prices

```python
import yfinance as yf
import matplotlib.pyplot as plt
```

```python
stock_ticker = 'AAPL'
start_date = '2024-01-01' # use format YYYY-MM-DD
end_date = '2024-12-31'

data = yf.download( stock_ticker, start= start_date, end = end_date ) 
```

```python
data.head()
```

```python
closing_prices = data['Close']
high_prices = data['High']
```

```python
# Plot a line chart for the closing prices
plt.figure(figsize=(10, 5))
plt.plot(closing_prices.index, closing_prices, label="AAPL Closing Prices", color='blue')
```

```python
# Plot a line chart for the closing prices
plt.figure(figsize=(10, 5))
plt.plot(closing_prices.index, closing_prices, label="AAPL Closing Prices", color='blue')
plt.plot(high_prices.index, high_prices, label="AAPL High Prices", color='red')
plt.title('Apple Closing Prices')
plt.xlabel('Date')
plt.ylabel('Closing Price (USD)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Plot a box-and-whisker plot for the closing prices
plt.figure(figsize=(4,6))
plt.boxplot(closing_prices, vert=True, patch_artist=True, boxprops=dict(facecolor='lightblue'))
plt.title('Apple Closing Price Distripution Jan-Dec 2024')
plt.ylabel('Closing Price (USD)')
plt.grid(True, which='both', axis='y', linestyle='--', linewidth=0.7)
plt.show()
```

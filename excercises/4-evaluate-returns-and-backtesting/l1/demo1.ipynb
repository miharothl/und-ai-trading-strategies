# Cumulative Compounded Returns

Assume you have invested $10.000 in a portfolio of stocks. Over the past five years the portfolio has generated the following annual returns 10%, -5%, 20%, 8%, -12%

```python
annual_returns = [0.1, -0.05, 0.2, 0.08, -0.12]
```

```python
annual_returns
```

```python
import numpy as np
```

```python
annual_returns = np.array(annual_returns)
```

```python
annual_returns
```

```python
cumulative_compounded_returns = np.cumprod(1 + annual_returns) - 1
```

```python
cumulative_compounded_returns
```

```python
cumulative_sum_of_returns = np.cumsum(annual_returns)
```

```python
cumulative_sum_of_returns
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(cumulative_compounded_returns, label = "Cumulative compound returns")
plt.plot(cumulative_sum_of_returns, label = "Cumulative sum of returns", linestyle="--")
plt.xlabel("Year")
plt.ylabel("Cumulative Return")
plt.legend()
plt.show()
```

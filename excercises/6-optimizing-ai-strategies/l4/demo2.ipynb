```python
import numpy as np
```

```python
# Set proportions
p_one = 0.54   # 54% will be 1
p_zero = 0.46  # 46% will be 0

# Generate random target distribution
n = len(data)
target = np.random.choice([1, 0], size=n, p=[p_one, p_zero])

# Add target column
data["target"] = target

print(data["target"].value_counts(normalize=True))
```

```python
# Drop rows with at least one NaN in any column
data = data.dropna()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(16, 14))
sns.heatmap(data.drop(columns=["Date", "Name"]).corr(), annot=False, cmap="coolwarm", center=0)
plt.show()
```

you need correlation between target and feature, in this case none as target is random

```python
X = data.drop(columns=["Date", "target", "Name"])
y = data["target"]

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, shuffle=False
)

X_train.shape, X_test.shape, y_train.shape, y_test.shape

from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators=5, max_depth=2, random_state=52)
model.fit(X_train, y_train)
```

```python
import matplotlib.pyplot as plt
from sklearn.model_selection import learning_curve
import numpy as np

def plot_learning_curve(model, X, y, cv=5, n_jobs=-1):
    train_sizes, train_scores, test_scores = learning_curve(
        model,
        X,
        y,
        cv=cv,
        n_jobs=n_jobs,
        verbose=0
    )
    
    train_scores_mean = np.mean(train_scores, axis=1)
    test_scores_mean = np.mean(test_scores, axis=1)

    plt.plot(train_sizes, train_scores_mean, label="CV training score")
    plt.plot(train_sizes, test_scores_mean, label="CV test score")
    plt.title("Learning curve for Random Forest Classifier")
    plt.xlabel("Training examples")
    plt.ylabel("Accuracy")
    plt.legend()
    plt.show()

plot_learning_curve(model, X_train, y_train)
```

```python
from sklearn.model_selection import cross_val_score

cross_val_score(model, X_train, y_train, cv=5, n_jobs=-1).mean()
```

```python
y_pred = model.predict(X_test)
acc_test = (y_pred == y_test).mean()
acc_test
```

```python
# inspect the features, which feature is most important
```

```python
from sklearn.inspection import permutation_importance

perm_imp_results = permutation_importance(
    model, X_test, y_test, n_repeats=10, n_jobs=-1, random_state=52
)
perm_importances = perm_imp_results.importances_mean
# indices = (-np.abs(perm_importances)).argsort()
indices = (perm_importances).argsort()

for i in range(X.shape[1]):
    print(f"{X.columns[indices[i]]}: {perm_importances[indices[i]]:.4f}")
```

```python
plt.figure(figsize=(10, 6))
plt.title("Feature Importance")
plt.bar(range(X.shape[1]), perm_importances[indices], align="center")
plt.xticks(range(X.shape[1]), X.columns[indices], rotation=90)
plt.xlabel("Features")
plt.ylabel("Importance")
plt.show()
```

```python
importances = model.feature_importances_
indices = (-importances).argsort()

for i in range(X.shape[1]):
    print(f"{X.columns[indices[i]]}: {importances[indices[i]]:.4f}")
```
